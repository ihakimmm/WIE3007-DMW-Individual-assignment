# Automated Feature Engineering

## Create EntitySets

Install Required Library

```python
pip install featuretools
```

Import Library

```python
import featuretools as ft
```

Display Datatype

```python
display(df_customers.dtypes)
display(df_orders.dtypes)
display(df_products.dtypes)
display(df_order_details.dtypes)
```

Create EntitySets

```python
es = ft.EntitySet('superstore')

# Add entities
es.add_dataframe(dataframe_name="customers",
                 dataframe=df_customers,
                 index="Customer ID")

es.add_dataframe(dataframe_name="orders",
                 dataframe=df_orders,
                 index="Order ID")

es.add_dataframe(dataframe_name="products",
                 dataframe=df_products,
                 index="Product ID")

es.add_dataframe(dataframe_name="order_details",
                 dataframe=df_order_details,
                 index='order_product_id')

es.add_relationship("customers", "Customer ID", "orders", "Customer ID")
es.add_relationship("products", "Product ID", "order_details", "Product ID")
es.add_relationship("orders", "Order ID", "order_details", "Order ID")
```

## Deep Feature Syntesis

#### Customers Table

```python
customers_features, customers_feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name='customers'
)
customers_features.reset_index(inplace=True)
```

Display Valid Features Generated

```python
# Identify columns with non-null and non-zero values
valid_customers_columns = customers_features.columns[(customers_features.notnull().all() & (customers_features != 0).all())]
```

List Valid Features Generated

```python
print(f"Total valid features for customers: {len(valid_customers_columns)}")
list(customers_features[valid_customers_columns].columns)
```

* Based on the features generated, we can select the most important and usable for the business.
1. COUNT(orders) as ordersCount
2. SUM(order_details.Sales) as totalSpent
3. SUM(order_details.Quantity) as totalQuantityPurchased

```python
customers_features = customers_features[['Customer ID', 'COUNT(orders)', 'SUM(order_details.Sales)', 'SUM(order_details.Quantity)']]
customers_features.rename(columns={'COUNT(orders)': 'ordersCount', 'SUM(order_details.Sales)': 'totalSpent', 'SUM(order_details.Quantity)': 'totalQuantityPurchased'}, inplace=True)
customers_features
```

```python
customers_features['Customer ID'] = customers_features['Customer ID'].astype('string')
display(customers_features.dtypes)
```

Merge the new generated features into Customers DataFrame

```python
df_customers = pd.merge(df_customers, customers_features, on='Customer ID', how='left')
df_customers.head()
```

Load the Customers table into BigQuery Datawarehouse

#### Orders table

```python
orders_features, orders_feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name='orders'
)
orders_features.reset_index(inplace=True)
```

```python
# Identify columns with non-null and non-zero values
valid_orders_columns = orders_features.columns[(orders_features.notnull().all() & (orders_features != 0).all())]
```

```python
print(f"Total valid features for customers: {len(valid_orders_columns)}")
list(orders_features[valid_orders_columns].columns)
```

Select the important and choosen features

```python
orders_features = orders_features[[
    'Order ID',
    'SUM(order_details.Sales)',
    'SUM(order_details.Quantity)',
    'DAY(Order Date)',
    'MONTH(Order Date)',
    'YEAR(Order Date)',
    'DAY(Ship Date)',
    'MONTH(Ship Date)',
    'YEAR(Ship Date)',
    'NUM_UNIQUE(order_details.products.Category)',
    'NUM_UNIQUE(order_details.products.Sub-Category)']]
orders_features.rename(columns={
    'SUM(order_details.Sales)': 'totalSales',
    'SUM(order_details.Quantity)': 'totalQuantity',
    'DAY(Order Date)': 'orderDay',
    'MONTH(Order Date)': 'orderMonth',
    'YEAR(Order Date)': 'orderYear',
    'DAY(Ship Date)': 'shipDay',
    'MONTH(Ship Date)': 'shipMonth',
    'YEAR(Ship Date)': 'shipYear',
    'NUM_UNIQUE(order_details.products.Category)': 'noCategory',
    'NUM_UNIQUE(order_details.products.Sub-Category)': 'noSubCategory',}, inplace=True)
orders_features.head()
```

```python
orders_features['Order ID'] = orders_features['Order ID'].astype('string')
display(orders_features.dtypes)
```

Merge with the original dataset

```python
df_orders = pd.merge(df_orders, orders_features, on='Order ID', how='left')
df_orders.head()
```

```python
# Convert columns to datetime objects
df_orders['orderDay'] = df_orders['orderDay'].astype('string')
df_orders['orderMonth'] = df_orders['orderMonth'].astype('string')
df_orders['orderYear'] = df_orders['orderYear'].astype('string')
df_orders['shipDay'] = df_orders['shipDay'].astype('string')
df_orders['shipYear'] = df_orders['shipYear'].astype('string')
df_orders['shipMonth'] = df_orders['shipMonth'].astype('string')

# Display updated data types
df_orders.dtypes
```
